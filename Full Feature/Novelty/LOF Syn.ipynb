```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_curve, auc, precision_score, recall_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import LocalOutlierFactor
from numpy import arange
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
```

```python
benign_df = pd.read_csv(r'D:\Kuliah\ProyekAkhir\New Data\resample\Benign_resample.csv')
benign_df.head(2)
```

```python
syn_df = pd.read_csv(r'D:\Kuliah\ProyekAkhir\New Data\resample\Syn_resample_4000.csv')
syn_df
```

```python
benign_test_df = pd.read_csv(r'D:\Kuliah\ProyekAkhir\New Data\resample\Benign_resample_4000.csv')
```

```python
benign_test_df
```

```python
df = pd.concat([benign_df, syn_df], ignore_index=True)
```

```python
df
```

```python
# novelty detection
df_test = pd.concat([benign_test_df, syn_df], ignore_index=True)
df_train = benign_df

# training_data
X_train = df_train.drop([' Label'], axis=1)
y_train = df_train[' Label']
# testing_data
X_test = df_test.drop([' Label'], axis=1)
y_test = df_test[' Label']
```

```python
df_test.to_csv(r'D:\Kuliah\ProyekAkhir\Data Weka\benign_syn.csv', index = False, header=True)
```

```python
# Outlier detection
X = df.drop([' Label'], axis=1)
y = df[' Label']
```

```python
y_test
```

```python
# Outlier Detection
for a in range(1, 100):
    for i in arange(0.01, 0.5, 0.01):
        #feature extraction
    #     pca = PCA(n_components=1)
    #     df_ddos_withpca = pca.fit_transform(df_ddos_normalization)
        #model
        clf = LocalOutlierFactor(n_neighbors=a, novelty=False, contamination=i)
        anomaly =clf.fit_predict(X)
        
        print('Neighbour : ' + str(a))
        print('contamination : ' + str(i))
        print('---Classification Report---')
        print(classification_report(y, anomaly))
```

```python
# Novelty Detection with 240 sample balanced
for a in range(1, 20):
    for i in arange(0.1, 0.5, 0.1):
        clf = LocalOutlierFactor(n_neighbors=a, novelty=True, contamination=i)
        clf.fit(X_train)
        y_pred_test = clf.predict(X_test)
        print('Neighbour : ' + str(a))
        print('contamination : ' + str(i))
        print('---Classification Report---')
        print(classification_report(y_test, y_pred_test))
```

```python
# Novelty detection with 8000 sample balanced
for a in range(1, 20):
    for i in arange(0.1, 0.5, 0.1):
        clf = LocalOutlierFactor(n_neighbors=a, novelty=True, contamination=i)
        clf.fit(X_train)
        y_pred_test = clf.predict(X_test)
        print('Neighbour : ' + str(a))
        print('contamination : ' + str(i))
        print('---Classification Report---')
        print(classification_report(y_test, y_pred_test))
```
